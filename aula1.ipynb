## Aula 1 Análise de dados, python, pandas e novos fármacos

```python
# importa o pandas e lê os dados em csv a partir de uma url
# importa matplotlib para os gráficos

import matplotlib.pyplot as plt
import pandas as pd

url_dados = 'https://github.com/alura-cursos/imersaodados3/blob/main/dados/dados_experimentos.zip?raw=true'

dados = pd.read_csv(url_dados, compression='zip')
```

```python
# mostra as primeiras 5 linhas do dataframe

dados.head()
```

```python
# plota um gráfico no formato pizza
# dados['tratamento'].value_counts().plot.pie()

# plota um gráfico no formato barra
dados['tratamento'].value_counts().plot.bar()
plt.title('Tratamento')
plt.xlabel('Tipo de tratamento')
```

### Desafio 01: Investigar porque a classe 'tratamento' é tão desbalanceada?


A classe 'tratamento' é desbalanceada por se tratar da classe que registra se as amostras receberam ou não o fármaco. As amostras que não recebem o fármaco são o grupo de controle da pesquisa. 

*Para saber mais*: [Grupo de controle](https://pt.wikipedia.org/wiki/Grupo_de_controle)

### Desafio 02: Mostrar as últimas linhas da tabela

### Desafio 03: Proporção das classes 'tratamento'.

```python
# pandas.Series.mul
# pandas.Series.round
dados['tratamento'].value_counts(normalize=True).mul(100).round(2)
```

```python
# mostra a proporção plotando um gráfico no formato pizza

dados_tratamento = dados['tratamento'].value_counts()
dados_tratamento.plot.pie()
```

### Desafio 05: Procurar na documentação o método query do pandas.



O método ***query*** é utilizado para filtrar os dados de um dataframe.

*Documentação*: [pandas.DataFrame.query](https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.query.html)

#### Exemplos

```python
# retorna linhas com valores maiores que 5 na coluna 'g-0'

dados.query('`g-0` > 5')
```

```python
# retorna linhas com valores iguais a 10 na coluna 'g-0'

dados.query('`g-0` == 10')
```
